```python
import matplotlib.pyplot as plt
import glob
```

```python
files = glob.glob('*.txt')
learning_rates = [0.0005, 0.001, 0.002]
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
for hlr in learning_rates:
    for llr in learning_rates:
        xs, ys = [], []
        with open(str(hlr) + "_" + str(llr) + ".txt") as f:
            for line in f:
                xy = line.strip().split('\t')
                x, y = int(xy[0]), float(xy[1])
                xs.append(x)
                ys.append(y)
        plt.plot(xs, ys, label=str(hlr) + "_" + str(llr))
ax.legend()
ax.set_xlabel('Number of iterations')
ax.set_ylabel('Average Accuracy across 4 subtasks')
plt.savefig('lr_performance.png');
```
